```python
import os
import pandas as pd
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.neighbors import NearestNeighbors
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model
```

```python
image_folder = "images"
image_files = [f for f in os.listdir(image_folder) if f.endswith('.jpg')]
```

```python
for i in range(0,4):
    img_path = os.path.join(image_folder, image_files[i])
    img = plt.imread(img_path)

    # Display the image
    plt.imshow(img)
    plt.title(f"Image {i+1}")
    plt.show()
```

```python
# Function to load and preprocess images using MobileNetV2
def load_and_preprocess_images_mobilenet(image_folder):
    images = []
    image_paths = []

    for filename in os.listdir(image_folder):
        img_path = os.path.join(image_folder, filename)

        try:
            img = Image.open(img_path)
        except (IOError, OSError) as e:
            print(f"Error loading image {img_path}: {e}")
            continue

        img = img.resize((224, 224))  # Resize to match the input size expected by MobileNetV2

        # Convert RGBA to RGB if the image has an alpha channel
        if img.mode == 'RGBA':
            img = img.convert('RGB')

        img_array = np.array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array = preprocess_input(img_array)
        images.append(img_array)
        image_paths.append(img_path)

    return np.vstack(images), image_paths

# Load and preprocess product images using MobileNetV2
product_images_mobilenet, image_paths_mobilenet = load_and_preprocess_images_mobilenet('images/')

# Load pre-trained MobileNetV2 model without the top (classification) layer
base_model_mobilenet = MobileNetV2(weights='imagenet', include_top=False, pooling='avg')

# Extract features using the MobileNetV2 model
features_mobilenet = base_model_mobilenet.predict(product_images_mobilenet)
```

```python
# Create a Nearest Neighbors model
model = NearestNeighbors(n_neighbors=5, metric='cosine')
model.fit(features_mobilenet)

# Function to find similar products based on user's input image
def find_similar_products(user_image_path, model, base_model_mobilenet):
    # Load and preprocess user's input image
    try:
        user_img = Image.open(user_image_path)
    except (IOError, OSError) as e:
        print(f"Error loading user's image {user_image_path}: {e}")
        return []

    user_img = user_img.resize((224, 224)) 
    if user_img.mode == 'RGBA':
        user_img = user_img.convert('RGB')  # Convert RGBA to RGB if the image has an alpha channel

    user_img_array = np.array(user_img)
    user_img_array = np.expand_dims(user_img_array, axis=0)
    user_img_array = preprocess_input(user_img_array)

    # Extract features for the user's input image
    user_features = base_model_mobilenet.predict(user_img_array)

    # Find the nearest neighbors based on cosine similarity
    _, indices = model.kneighbors(user_features)

    return indices[0]
```

```python
# Path to the user's input image
user_image_path = 'images/image272.jpg'

# Find similar products based on the user's input image
similar_products_indices = find_similar_products(user_image_path, model, base_model_mobilenet)

# Display the user's input image
user_img = Image.open(user_image_path)
plt.imshow(user_img)
plt.title("User's Input Image")
plt.show()

# Display the top 5 similar product images
for idx in similar_products_indices:
    similar_product_img = Image.open(image_paths_mobilenet[idx])
    plt.imshow(similar_product_img)
    plt.title(f"Similar Product {idx + 1}")
    plt.show()
```
